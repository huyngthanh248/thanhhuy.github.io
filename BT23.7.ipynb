```python
import math
import turtle

xa = xA = int(input("xA = "))
ya = yA = int(input("yA = "))
xb = xB = int(input("xB = "))
yb = yB = int(input("yB = "))
xc = xC = int(input("xC = "))
yc = yC = int(input("yC = "))


AB = BA = math.sqrt((xb-xa)**2 + (yb-ya)**2)
AC = CA = math.sqrt((xc-xa)**2 + (yc-ya)**2)
BC = CB = math.sqrt((xc-xb)**2 + (yc-yb)**2)


tamgiac = AB+AC>BC and AB+BC>AC and AC+BC>AB
print("Đây là tam giác", tamgiac)


cosA = ((AB**2 + AC**2 - BC**2)/(2 * AB * AC))
cosB = ((BA**2 + BC**2 - AC**2)/(2 * BA * BC))
cosC = ((CB**2 + CA**2 - BA**2)/(2 * CB * CA))
#đường cao
p = (AB + BC + CA)/2
HA= math.sqrt(p*(p - AB)*(p - BC)*(p - AC))/BC
HB= math.sqrt(p*(p - AB)*(p - BC)*(p - AC))/AC
HC= math.sqrt(p*(p - AB)*(p - BC)*(p - AC))/AB

print("Độ dài đường cao từ đỉnh A là: " + str(HA))
print("Độ dài đường cao từ đỉnh B là: " + str(HB))
print("Độ dài đường cao từ đỉnh C là: " + str(HC))

#trung điểm
#AB
xm = xM = (xa + xb)/2
ym = yM = (ya + yb)/2 
#AC
xn = xN = (xa + xc)/2
yn = yN = (ya + yc)/2
#BC
xo = xO = (xb + xc)/2
yo = yO = (yb + yc)/2
#độ dài trung tuyến
CM = MC = math.sqrt((xm-xc)**2 + (ym-yc)**2)
BN = NB = math.sqrt((xn-xb)**2 + (yn-yb)**2)
AO = OA = math.sqrt((xo-xa)**2 + (yo-ya)**2)

# Print độ dài trung tuyến
print("Độ dài đường trung tuyến từ A (AO) là: " + str(OA))
print("Độ dài đường trung tuyến từ B (BN) là: " + str(BN))
print("Độ dài đường trung tuyến từ C (CM) là: " + str(CM))


# Tọa độ trọng tâm
xi = xI = (xa + xb + xc)/3
yi = yI = (ya + yb + yc)/3
print("Tọa độ trọng tâm của tam giác ABC là: (" + str(xi) + "; " + str(yi) + ")")

#tọa độ trực tâm
#vẽ
t=turtle.Turtle()
t.shape("turtle")
t.pensize(1)

t.goto(xA, yA)
t.pendown()
t.goto(xB, yB)
t.goto(xC, yC)
t.goto(xA, yA)
t.penup()
t.goto(xA, yA)
t.pendown()
t.goto(xO, yO)
t.penup()
t.goto(xB, yB)
t.pendown()
t.goto(xN, yN)
t.penup()
t.goto(xC, yC)
t.pendown()
t.goto(xM, yM)
t.penup()
t.goto(xI, yI)
t.pendown()

turtle.done
```
