```python
import googletrans
# print(googletrans.LANGUAGES)
from googletrans import Translator
from tkinter import *
import pyttsx3
from tkinter import ttk, messagebox
import textblob

window = Tk()
window.title("Google Translate")
window.geometry("400x500")
window.resizable(False,False)

language =(googletrans.LANGUAGES)
language_list = list(language.values())

def trans():
    text_output.delete(1.0, END)
    try:
        for key, values in language.items():
            if (values == combo1.get()):
                from_lang_key = key

        for key, values in language.items():
            if (values == combo2.get()):
                to_lang_key = key
    
        
        words = textblob.TextBlob(text_input.get(1.0,END))
        words = words.translate(from_lang= from_lang_key,to = to_lang_key)
        text_output.insert(1.0, words)
    
    except Exception as e:
        messagebox.showerror("Translator",e)
    
    
        

def clear():
    text_input.delete(1.0,END)
    text_output.delete(1.0,END)
def speech():
    engine = pyttsx3.init()
    engine.say(text_output.get(1.0,END))
    engine.runAndWait()
    text_output.delete(1.0,END)
#     pass
    
lb1 = Label(window, text="Google translate", font='arial 20', fg='black')
lb1.place(x=90,y=20)
##input
frame_input = LabelFrame(window,text ="Input",font='arial 16')
frame_input.place(x=40,y=70)

combo1 = ttk.Combobox(frame_input, width=30, value=language_list)
combo1.current("21")
combo1.pack()

text_input = Text(frame_input, width=45, height = 6, font='arial 12')
text_input.pack()
##output
frame_output = LabelFrame(window,text ="Output",font='arial 16')
frame_output.place(x=40,y=200)

combo2 = ttk.Combobox(frame_output, width=30, value=language_list)
combo2.pack()

text_output = Text(frame_output, width=45, height = 6, font='arial 12')
text_output.pack()


##button
btn_trans = Button(window,text="Translate",width=5,command = trans)
btn_trans.place(x=50,y=350)

btn_clear = Button(window,text="Clear all",width=5,command = clear)
btn_clear.place(x=150,y=350)

btn_speech = Button(window,text="Speech",width=5,command = speech)
btn_speech.place(x=250,y=350)


#Speech



window.mainloop()
```
